```python
import numpy as np
import matplotlib.pyplot as plt
data_path='/home/irmak/Workspaces/lectures_ws/lectures_git/cmpe_492/human_motion_modelling/data/prediction/'
propagation_steps = [3,10,20]
```

```python
# visualize right hand training and prediction errors
_, axs = plt.subplots(1,2, figsize=(14,3))
for p in propagation_steps:
    training_errors = np.load(open(data_path+'training_errors_right_hand_only_'+str(p)+'_prop_step.npy','rb'))
    prediction_errors = np.load(open(data_path+'prediction_errors_right_hand_only_'+str(p)+'_prop_step.npy','rb'))
    axs[0].plot(range(len(training_errors)), training_errors, label=str(p)+' prop steps')    
    axs[1].plot(range(len(prediction_errors)), prediction_errors, label=str(p)+' prop steps')

axs[0].set_ylabel('Errors with only \nRight Hand given', fontsize=12)
axs[0].set_title('Training Errors')
axs[1].set_title('Prediction Errors')
axs[0].legend()
axs[1].legend()

# # visualize right hand and left foot training and prediction errors
# _, axs = plt.subplots(1,2, figsize=(15,7))
# for p in propagation_steps:
#     training_errors = np.load(open(data_path+'training_errors_right_hand_left_foot_'+str(p)+'_prop_step.npy','rb'))
#     prediction_errors = np.load(open(data_path+'prediction_errors_right_hand_left_foot_'+str(p)+'_prop_step.npy','rb'))
#     axs[0].plot(range(len(training_errors)), training_errors, label=str(p)+' prop steps')    
#     axs[1].plot(range(len(prediction_errors)), prediction_errors, label=str(p)+' prop steps')

# axs[0].set_ylabel('Errors with Right Hand & Left Foot given', fontsize=15)
# axs[0].set_title('Training Errors')
# axs[1].set_title('Prediction Errors')
# axs[0].legend()
# axs[1].legend()

# visualize right left hand and left foot training and prediction errors
_, axs = plt.subplots(1,2, figsize=(14,3))
for p in propagation_steps:
    training_errors = np.load(open(data_path+'training_errors_right_left_hand_left_foot_'+str(p)+'_prop_step.npy','rb'))
    prediction_errors = np.load(open(data_path+'prediction_errors_right_left_hand_left_foot_'+str(p)+'_prop_step.npy','rb'))
    axs[0].plot(range(len(training_errors)), training_errors, label=str(p)+' prop steps')    
    axs[1].plot(range(len(prediction_errors)), prediction_errors, label=str(p)+' prop steps')

axs[0].set_ylabel('Errors with Right & Left Hand &\nLeft Foot given', fontsize=12)
axs[0].set_title('Training Errors')
axs[1].set_title('Prediction Errors')
axs[0].legend()
axs[1].legend()

# visualize head,hands&feet training and prediction errors
_, axs = plt.subplots(1,2, figsize=(14,3))
for p in propagation_steps:
    training_errors = np.load(open(data_path+'training_errors_head_hands_feet_'+str(p)+'_prop_step.npy','rb'))
    prediction_errors = np.load(open(data_path+'prediction_errors_head_hands_feet_'+str(p)+'_prop_step.npy','rb'))
    axs[0].plot(range(len(training_errors)), training_errors, label=str(p)+' prop steps')    
    axs[1].plot(range(len(prediction_errors)), prediction_errors, label=str(p)+' prop steps')

axs[0].set_ylabel('Errors with Head & Hands &\nFeet given', fontsize=12)
axs[0].set_title('Training Errors')
axs[1].set_title('Prediction Errors')
axs[0].legend()
axs[1].legend()
```

```python
# show how each of the joints' positions and the predictions change in timesteps, for 389. trajectory with only right hand given
wanted_objects = np.load(open(data_path+'wanted_objects.npy', 'rb'))
predicted_objects = np.load(open(data_path+'predicted_objects_right_hand_10.npy', 'rb'))

# wanted_joints = [0, 1, 3,5,6, 7,9,10, 11, 13,14,16, 17,18,20]
wanted_joints = [13,14,16]
# joint_label = [
#     'STMCH',
#     'CHEST',
#     'LSLD',
#     'LELB',
#     'LHND',
#     'RSLD',
#     'RELB',
#     'RHND',
#     'HEAD',
#     'LHIP',
#     'LKNEE',
#     'LFOOT',
#     'RHIP',
#     'RKNEE',
#     'RFOOT',
# ]
joint_label = [
    'LHIP',
    'LKNEE',
    'LFOOT',
]

print('wanted_objects.shape: {}, predicted_objects.shape: {}'.format(wanted_objects.shape, predicted_objects[:,:,wanted_joints,:].shape))

frame_length = 795

row = predicted_objects[:,:,wanted_joints,:].shape[2]
col = predicted_objects.shape[3]
fig, axs = plt.subplots(row,
                        col,
                        figsize=(col*5, row*3))
for l,i in enumerate(wanted_joints):
    for j in range(predicted_objects.shape[3]):
        axs[l,j].plot(range(len(predicted_objects[1,:frame_length,i,j])), predicted_objects[1,:frame_length,i,j])
        axs[l,j].plot(range(len(predicted_objects[1,:frame_length,i,j])), wanted_objects[1,1:1+frame_length,i,j])
        axs[l,j].get_xaxis().set_ticks([])
    axs[l,0].set_ylabel(joint_label[l])
axs[0,1].set_title("Links of left leg with right hand given as the reference points", fontsize=15)
```

```python
# subset of the plots above that shows the joints when right hand is given as the reference point
# it will show the joints, LFOOT,LKNEE, RHND and RELB and with 0,2nd joints only (z,y)

wanted_objects = np.load(open(data_path+'wanted_objects.npy', 'rb'))
predicted_objects = np.load(open(data_path+'predicted_objects_right_hand_10.npy', 'rb'))

# wanted_joints = [0, 1, 3,5,6, 7,9,10, 11, 13,14,16, 17,18,20]
wanted_joints = [9,14,10,16]
# joint_label = [
#     'STMCH',
#     'CHEST',
#     'LSLD',
#     'LELB',
#     'LHND',
#     'RSLD',
#     'RELB',
#     'RHND',
#     'HEAD',
#     'LHIP',
#     'LKNEE',
#     'LFOOT',
#     'RHIP',
#     'RKNEE',
#     'RFOOT',
# ]
joint_label = [
    'RELB',
    'LKNEE',
    'RHND',
    'LFOOT',
]

# print('wanted_objects.shape: {}, predicted_objects.shape: {}'.format(wanted_objects.shape, predicted_objects[:,:,wanted_joints,:].shape))

frame_length = 795


row = 2
col = 4

# row = predicted_objects[:,:,wanted_joints,:].shape[2]
# col = predicted_objects.shape[3]
fig, axs = plt.subplots(row,
                        col,
                        figsize=(col*7, row*4),
                        constrained_layout=True)


for i in range(row):
    for j in range(col):
        
        # set the joint_index to draw
        joint_enum = i*2 + (j >= 2)
        joint_index = wanted_joints[joint_enum]
        print('(i,j): {},{}, joint_index: {}'.format(i,j,joint_index))
    
    
        # set the axis to draw and set the labels
        if j % 2 == 0: # j is 0 or 2
            axis = 0
            axs[i,j].set_ylabel(joint_label[joint_enum], fontsize=30)
        else:
            axis = 2
    
        if i == 0 and j == 0:
            axs[i,j].plot(range(frame_length), predicted_objects[1,:frame_length,joint_index,axis], '-', linewidth=4, label="Predictions")
            axs[i,j].plot(range(frame_length), wanted_objects[1,1:1+frame_length,joint_index,axis], '--', linewidth=4, label="Ground Truth")
            axs[i,j].legend(prop={"size":25})
        else:
            axs[i,j].plot(range(frame_length), predicted_objects[1,:frame_length,joint_index,axis], '-', linewidth=4)
            axs[i,j].plot(range(frame_length), wanted_objects[1,1:1+frame_length,joint_index,axis], '--', linewidth=4)
        axs[i,j].get_xaxis().set_ticks([])
        axs[i,j].xaxis.set_tick_params(labelsize=20)
        axs[i,j].yaxis.set_tick_params(labelsize=20)
```
